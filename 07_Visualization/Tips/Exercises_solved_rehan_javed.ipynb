### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

% matplotlib inline
```

```python
import plotly.express as px
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
address = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips=pd.read_csv(address)
tips.head(10)
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']

tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(data=tips, x='total_bill')
```

```python
fig = px.histogram(tips, x="total_bill", nbins=15)
fig.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = tips)
```

```python
px.scatter(data_frame=tips, x='total_bill', y='tip', width=1000, height=600)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
tips_2 = tips[["total_bill", "tip", "size"]]
```

```python
px.scatter_matrix(data_frame=tips_2,width=1000, height=1000)
```

```python
sns.pairplot(tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
px.box(data_frame=tips, x='day', y='total_bill')
```

```python
px.strip(data_frame=tips, x='day', y='total_bill')
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(data=tips, x='tip', y='day', hue='sex', jitter=True)
```

```python
px.strip(data_frame=tips, x='tip', y='day', width=1000, height=600, color='sex')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
```

```python
px.box(data_frame=tips, x='day', y='total_bill', color='time')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.set(style="ticks")
g = sns.FacetGrid(data=tips, col='time')
g.map(plt.hist, 'tip')
```

```python
px.histogram(data_frame=tips, x='tip',facet_col='time')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g1 = sns.FacetGrid(data=tips, col='time', hue='smoker')
g1.map(plt.scatter, 'total_bill', 'tip')
```

```python
px.scatter(data_frame=tips, x='total_bill', y='tip', color='smoker', facet_col='sex')
```
