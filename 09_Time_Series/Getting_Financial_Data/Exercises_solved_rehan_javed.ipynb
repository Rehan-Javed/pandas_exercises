### Step 8. Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
all.head()
```

```python
all['year'] = all['date'].dt.year
```

```python
all.head(2)
```

```python
all['week'] = all['date'].dt.isocalendar().week
```

```python
all.head()
```

```python
all.shape
```

```python
weekly_data = all.groupby(by=['year', 'week', 'stock']).agg({
    '5. volume' : 'sum'
})
```

```python
weekly_data.shape
```

```python
weekly_data.reset_index(inplace=True)
```

```python
weekly_data
```

```python
pd.pivot_table(weekly_data, values = '5. volume', index=['year', 'week'],
              columns=['stock'], aggfunc=np.sum)
```
