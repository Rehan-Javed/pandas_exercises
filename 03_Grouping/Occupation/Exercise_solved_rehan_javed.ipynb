### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(address, sep='|')
```

```python
users.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
users_occ_gp = users.groupby(by='occupation').agg({
    'age' : 'mean'

})
```

```python
users_occ_gp.reset_index(inplace=True)
```

```python
users_occ_gp
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def is_male_func(x):
    if x == 'M':
        return 1
    elif x!= 'M':
        return 0

users['is_male'] = users['gender'].apply(is_male_func)
```

```python
#using map

gender_map = {'M' : 1, 'F' : 0}

users['is_male_mapped'] = users.gender.map(gender_map)
```

```python
# using np.where
import numpy as np
```

```python
users['is_male_where'] = np.where(users['gender'] == 'M', 1, 0)
```

```python
users.head()
```

```python
ratio = users.groupby(by='occupation').agg({
    'is_male' : 'sum'
})
```

```python
ratio['total'] = users.occupation.value_counts()
```

```python
ratio.reset_index(inplace=True)
```

```python
ratio.head()
```

```python
ratio['male_ratio'] = ratio.is_male / ratio.total
```

```python
ratio.sort_values('male_ratio', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
age = users.groupby(by='occupation').age.agg(['min', 'max'])
```

```python
age.reset_index(inplace=True)
```

```python
age
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
occu_gend = users.groupby(by=['occupation', 'gender']).age.agg(['mean','min', 'max'])
```

```python
occu_gend
```

### Step 8.  For each occupation present the percentage of women and men

```python
ratios = ratio[['occupation', 'male_ratio']]
```

```python
ratios['male_percentage'] = ratios.loc[:, 'male_ratio']*100
```

```python
ratios['female_percentage'] = 100 - ratios.loc[:,'male_percentage']
```

```python
ratios.drop('male_ratio', inplace=True, axis=1)
```

```python
ratios
```
